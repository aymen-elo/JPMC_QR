```python
# Helper functions
def linear_interpolate(A, B, X, a, b):
    return ((b - a)/(B - A).days)*(X - A).days + a

def extract_date(string_date):
    return dt.datetime.strptime(string_date, "%m/%d/%y")
```

```python
datestr = "8/10/24"
date = extract_date(datestr)

min_date = extract_date(df.iloc[0].iloc[0])
max_date = extract_date(df.iloc[-1].iloc[0])
```

```python
def price(date):

    price = 0

    # Check if date is in or out range:
    is_in_range = date < max_date and date > min_date
    
    if(is_in_range):
        # price inside range - interpolation
        prev_date = date.replace(day=1) - dt.timedelta(days=1)
        prev_price = float(df.loc[df['Dates'] == f"{prev_date.month}/{prev_date.day}/{prev_date.year - 2000}", 'Prices'].values[0]) 
        
        next_date = date.replace(day=1, month=date.month+1) - dt.timedelta(days=1)
        next_price = float(df.loc[df['Dates'] == f"{next_date.month}/{next_date.day}/{next_date.year - 2000}", 'Prices'].values[0]) 
        
        price = linear_interpolate(prev_date, next_date, date, prev_price, next_price)
    else:
        # price beyond range - extrapolation
        ""
    
    return price
```

```python
price(date)
```
